## Web Scrapping de las tendencias de Twitter
Con este c√≥digo sacaremos los hastaghs que han sido <i>Trending toppic</i> <b>mundial</b> y en <b>Estados Unidos</b> por franjas horarias diarias. 

```python
import requests as r
from bs4 import BeautifulSoup
import json
```

```python

horas = [x for x in range(24)]
meses = {"01":31, "02":28, "03":31, "04":30, "05":31, "06":30, "07":31, "08":31, "09":30, "10":31, "11":30, "12":31}
respuestas = []
calendario = {}
for mes in meses:
    dias = ["0"+str(dia+1) for dia in range(meses[mes]) if dia <10]
    calendario[mes] = {dia:horas for dia in dias}
    

print(json.dumps(calendario, indent = 4))  # ver el diccionario como json indentado

```

```python
# sin esta informaci√≥n de cabecera da error 403
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.113 Safari/537.36'}

for mes in meses:
    print(f"Vamos por el mes: {mes}")
    for dia in range(meses[mes]):
        print(f"Vamos por el d√≠a: {dia}")
        for hora in range(24):
            respuestas.append(r.get(f"https://getdaytrends.com/es/2021-{mes}-{dia}/{hora}/", headers=headers))
print(respuestas)
```

```python

for mes in respuestas:
    for dia in mes:
        for respuesta in dia:
# hacemos sopa con la respuesta
            sopa = BeautifulSoup(respuesta.content)
            tabla = sopa.find("table")
            tt = []
            for t in tabla.find_all("a"):
                tt.append(t.text)
            print(tt)


```
